```julia
Pkg.add("Lathe")
```

```julia
using DataFrames
df = DataFrame(:X => randn(500), :Y => randn(500))
```

```julia
using Lathe.preprocess: StandardScalar
```

```julia
# It is a bit silly to need to fit this twice,
#     This should be changed in 0.1.2:
scaler = StandardScalar(df[!, :X])
```

```julia
println(scaler.predict(df[!, :X]))
```

```julia
scaled = scaler.predict(df[!, :X])
```

```julia
df[:trainX] = scaled
```

```julia
show(df)
```

```julia
using Lathe.models: LinearRegression
```

```julia
model = LinearRegression(df[:trainX], df[:Y])
```

```julia
yhat = model.predict()
```
